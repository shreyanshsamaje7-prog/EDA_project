```python
train=pd.read_csv("/content/sample_data/train.csv")
train.sample(15)
```

```python
plt.style.use("ggplot")
servived=train[train['Survived']==1].count()[0]

non_servived=train[train['Survived']==0].count()[0]
data=[servived,non_servived]
labels=["servived","non-servived"]
plt.pie(data,labels=labels,autopct="%.2f %%")
plt.legend()
plt.show()
```

```python

child=train.loc[train.Age < 15].count()['PassengerId']
adult=train.loc[(train.Age >= 15) & (train.Age)<31].count()['PassengerId']
senior=train.loc[(train.Age >= 31) & (train.Age)<64].count()['PassengerId']
senior_adult=train.loc[(train.Age >64) ].count()['PassengerId']
data=[child,adult,senior,senior_adult]
bins=['child','adult','senior','older']

plt.pie(data,labels=bins,autopct="%.2f %%",explode=[0.2,0,0,.3])

plt.show()


```

```python
male=train.loc[(train.Sex =='male') & (train.Survived==1)].count()['PassengerId']
plt.style.use('default')
female=train.loc[(train.Sex =='female') & (train.Survived==1)].count()['PassengerId']
data=[male,female]
labels=['male','female']
plt.bar(labels,data)
plt.yticks(np.arange(0,300,20))
plt.show()
```

```python
plt.style.use('default')
cls1=train.loc[(train.Pclass==1) & (train.Survived==1)].count()['PassengerId']
cls2=train.loc[(train.Pclass==2) & (train.Survived==1)].count()['PassengerId']
cls3=train.loc[(train.Pclass==3) & (train.Survived==1)].count()['PassengerId']

data=[cls1,cls2,cls3]
labels=['cls1','cls2','cls3']
plt.pie(data,labels=labels,autopct="%.2f %%")
plt.title("survival rate")
plt.legend()


plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
netflix=pd.read_csv('/content/sample_data/netflix_titles.csv')
netflix.tail(10)
```

```python
netflix.groupby('release_year').count()['show_id'].sort_values(ascending=False).head(1)
```

```python
flix=netflix.copy()
flix.head()
flix['ad_year']=pd.to_datetime(flix['date_added'],errors='coerce').dt.year
flix.head()
flix.groupby('ad_year').count()['title']
```

```python
netflix[netflix['type']=='Movie']['duration'].str.split(' ').str[0].astype('float').mean()
```

```python
flix['listed_in']=flix['listed_in'].str.split(', ')

gener_df=flix.explode('listed_in')

gener_lead=gener_df.groupby(['ad_year','listed_in']).size().reset_index(name='count')

gener_pivot=gener_lead.pivot(index='ad_year', columns='listed_in', values='count')

```

```python
movies = flix[flix['type'] == 'Movie']

movies['duration_min'] = movies['duration'].str.replace(' min','').astype(float)

movies['listed_in'] = movies['listed_in'].str.split(', ')

movie_genre = movies.explode('listed_in')

genre_duration = movie_genre.groupby('listed_in')['duration_min'].mean().sort_values(ascending=False)

genre_duration.head(10)
```

```python
flix['cast'].str.split(', ').explode().value_counts().sort_values(ascending=False).head(10)
```

```python
df=flix.dropna(subset=['cast'])

df['cast']=df['cast'].str.split(', ')

from itertools import combinations

collab=[]

for actors in df['cast']:
  collab.extend(combinations(actors,2))

collab_df=pd.DataFrame(collab,columns=['Actor1','Actor2'])

collab_df.value_counts().reset_index(name='count').sort_values(by='count',ascending=False).head(10)
```

```python
df = flix.dropna(subset=['country','listed_in'])

df['listed_in'] = df['listed_in'].str.split(', ')
df['country'] = df['country'].str.split(',')

gen_df = df.explode('listed_in')
gen_df = gen_df.explode('country')

gen_df['country'] = gen_df['country'].str.strip()

genre_country = gen_df.groupby(['country','listed_in']).size().reset_index(name='count').sort_values('count',ascending=False)

genre_country.head()
```
